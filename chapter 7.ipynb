```python
#                           CONTINUE STATMENT
# "continue" use to stop the currenrt iteration of the loop and continue with next one it instructs the program to "skip the iteration"
```

```python
for i in range(10):
    if i == 5:
        continue ###### it skip one value from the list
    print(i)
```

```python
## even number 
for i in range(10):
    if i == 2:
        continue
    print(i)
```

```python
# find given number is prime or not
a = int(input("enter number: "))
prime = False
for i in range (2, a):
    if (a%i == 0):
        prime = True
        break
if prime:
    print("this number is prime ")
else:
    print("given number is not prime number ")
```
